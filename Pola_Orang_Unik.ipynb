```python
# ===========================================
# Menampilkan Gambar Orang dengan Python
# ===========================================

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import urllib.request
from google.colab import files

# ---------------------------
# 1. Upload gambar sendiri
# ---------------------------
print("ðŸ‘‰ Silakan upload file gambar dari komputer (jpg/png):")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

img = Image.open(filename)
plt.figure(figsize=(4,4))
plt.imshow(img)
plt.axis("off")
plt.title("Gambar Orang (Upload Sendiri)")
plt.show()


# ---------------------------
# 2. Gambar dari internet
# ---------------------------
url = "https://images.unsplash.com/photo-1508214751196-bcfd4ca60f91"
urllib.request.urlretrieve(url, "person.jpg")

img2 = Image.open("person.jpg")
plt.figure(figsize=(4,4))
plt.imshow(img2)
plt.axis("off")
plt.title("Gambar Orang (Dari Internet)")
plt.show()


# ---------------------------
# 3. Gambar wajah sederhana dengan NumPy
# ---------------------------
img3 = np.ones((300, 300, 3))

# Bentuk wajah (lingkaran kuning)
for i in range(300):
    for j in range(300):
        if (i-150)**2 + (j-150)**2 < 100**2:
            img3[i, j] = [1, 1, 0]  # kuning

# Mata
img3[120:130, 120:130] = [0, 0, 0]
img3[120:130, 170:180] = [0, 0, 0]

# Mulut
img3[200:210, 120:180] = [1, 0, 0]

plt.figure(figsize=(4,4))
plt.imshow(img3)
plt.axis("off")
plt.title("Gambar Orang (Abstrak dengan NumPy)")
plt.show()
```
