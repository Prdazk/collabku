```python
#@title contoh tanaman jagung
import matplotlib.pyplot as plt
import numpy as np

# Data simulasi
x = np.linspace(0, 30, 200)  # kecepatan angin
y = -0.02*(x-12)**2 + 6      # pola melengkung (naik lalu turun)

# Plot grafik
plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.xlabel("Kecepatan Angin")
plt.ylabel("Hasil Panen Jagung")
plt.title("Pola Melengkung (Contoh untuk Polynomial Regression)")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data contoh: kecepatan pelari tiap interval
X = np.array([0, 1, 2, 3, 4, 5]).reshape(-1, 1)  # misal menit ke-0 sampai 5
Y = np.array([10, 14, 17, 18, 17, 16])           # kecepatan km/jam

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
Y_pred_linear = lin_reg.predict(X)

# Polynomial Regression (derajat 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, Y)
Y_pred_poly = poly_reg.predict(X_poly)

# Plot
plt.scatter(X, Y, color='red', label='Data Pelari')
plt.plot(X, Y_pred_linear, color='blue', label='Linear Regression')
plt.plot(X, Y_pred_poly, color='green', label='Polynomial Regression (xÂ²)')
plt.title("Linear vs Polynomial Regression - Kecepatan Pelari")
plt.xlabel("Waktu (menit)")
plt.ylabel("Kecepatan (km/jam)")
plt.legend()
plt.grid(True)
plt.show()
```
