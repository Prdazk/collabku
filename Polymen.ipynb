```python
#@title Hipotesa Function
# Import library
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data: jarak (X) dan kecepatan (Y)
X = np.array([0, 100, 200, 300, 400]).reshape(-1,1)
Y = np.array([10, 14, 17, 18, 17])

# Ubah fitur menjadi polinomial (derajat 2)
X_poly = PolynomialFeatures(degree=2).fit_transform(X)

# Buat model dan latih
model = LinearRegression()
model.fit(X_poly, Y)

# Prediksi
Y_pred = model.predict(X_poly)

# Tampilkan hasil
plt.scatter(X, Y, color='blue', label='Data Asli')
plt.plot(X, Y_pred, color='red', label='Prediksi')
plt.xlabel('Jarak (m)')
plt.ylabel('Kecepatan (km/jam)')
plt.title('Polynomial Regression - Kecepatan Pelari')
plt.legend()
plt.show()

```

```python
#@title Cost Function
# Import library
from sklearn.metrics import mean_squared_error

# Data contoh: nilai asli dan prediksi
nilai_asli = [90, 85, 88]        # nilai asli siswa
prediksi_model = [85, 80, 90]    # prediksi model

# Hitung MSE
mse = mean_squared_error(nilai_asli, prediksi_model)
print(f"(MSE): {mse}")

for i in range(len(nilai_asli)):
    selisih = nilai_asli[i] - prediksi_model[i]
    kuadrat = selisih**2
    print(f"Siswa {i+1}: Nilai asli = {nilai_asli[i]}, Prediksi = {prediksi_model[i]}, Selisih = {selisih}, Kuadrat = {kuadrat}")
```

```python
#@title Kelebihan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data: jarak vs kecepatan pelari
X = np.array([0,100,200,300,400]).reshape(-1,1)  # fitur (jarak)
Y = np.array([10,14,17,18,17])                   # target (kecepatan)

# Linear Regression: prediksi garis lurus
Y_lin = LinearRegression().fit(X,Y).predict(X)

# Polynomial Regression derajat 2: prediksi kurva mengikuti data
Y_poly = LinearRegression().fit(PolynomialFeatures(2).fit_transform(X),Y).predict(PolynomialFeatures(2).fit_transform(X))

# Visualisasi: titik data asli dan prediksi Linear vs Polynomial
plt.scatter(X,Y,color='blue',label='Data Asli')     # titik data asli
plt.plot(X,Y_lin,color='green',label='Linear')      # garis Linear
plt.plot(X,Y_poly,color='red',label='Polynomial')   # kurva Polynomial
plt.xlabel('Jarak (m)')
plt.ylabel('Kecepatan (km/jam)')
plt.legend()
plt.show()
```

```python
#@title memilih derajat polinomial
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = np.array([0, 100, 200, 300, 400]).reshape(-1,1)
Y = np.array([10, 14, 17, 18, 17])

# Coba beberapa derajat polinomial: 1=Linear, 2, 3
degrees = [1, 2, 3]
colors = ['green', 'red', 'purple']

for d, c in zip(degrees, colors):
    X_poly = PolynomialFeatures(d).fit_transform(X)  # tambah pangkat X
    Y_pred = LinearRegression().fit(X_poly, Y).predict(X_poly)  # prediksi
    plt.plot(X, Y_pred, color=c, label=f'Degree {d}')

# Titik data asli
plt.scatter(X, Y, color='blue', label='Data Asli')
plt.xlabel('Jarak (m)')
plt.ylabel('Kecepatan (km/jam)')
plt.title('Pengaruh Derajat Polinomial')
plt.legend()
plt.show()
```

```python
#@title Contoh kasus
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data: kecepatan vs durasi bertahan
X = np.array([8, 10, 12, 14, 16]).reshape(-1,1)
Y = np.array([45, 43, 35, 25, 15])

# Polynomial Regression derajat 2
X_poly = PolynomialFeatures(2).fit_transform(X)       # tambah X^2
Y_poly = LinearRegression().fit(X_poly, Y).predict(X_poly)  # prediksi kurva

# Visualisasi
plt.scatter(X, Y, color='blue', label='Data Asli')   # titik data asli
plt.plot(X, Y_poly, color='red', label='Polynomial') # kurva Polynomial
plt.xlabel('Kecepatan (km/j)')
plt.ylabel('Durasi Bertahan (menit)')
plt.legend()
plt.show()
```
