```python
import torch
import torch.nn as nn
import torch.optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, TensorDataset
```

```python
#@title Data numpy

x = torch.randn(100,2)
y = 3*x[:,0] + 2*x[:,1] + 0.1*torch.randn(100)

y = y.view(-1,1)

print(x)
print(y)
```

```python
dataset= TensorDataset(x,y)
dataloader = DataLoader(
    dataset,batch_size=10,shuffle=True
)
```

```python
from torch.autograd import forward_ad
class SimpleNN(nn.model):
  def __init__(self):
    super(SimpleNN, self).__init__()
    self.net = nn.Sequential(
        nn.Linear(2,8),
        nn.Relu(),
        nn.Linear(8,4),
        nn.Relu(),
        nn.Linear(4,1)
    )
  def forward(self, x):
      return self.net(x)

  model = SimpleNN()
  print(model)

```

```python
from torchsummary import summary
summary(model, (2,))
```

```python
criteria = nn.MSELoss()
optimazer = torch.optim.Adam(
    model.parameters(), lr=0.01)
for epoch in range(100):
  for inputs, targets in dataloader:
    outputs = model(inputs)
    loss = criteria(outputs, targets)
    optimazer.zero_grad()
    loss.backward()
    optimazer.step()
    if (epoch+1) % 10 == 0:
      print(f" Epoch [{epoch+1}/100], loss: {loss.item():.4f}")
```

```python
data_test= torch.tensor([[2.0,1.0]])
prediksi = model(data_test)
print(prediksi)
```
