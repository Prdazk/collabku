```python
from ultralytics import YOLO

model = YOLO("yolov8n.yaml")  # model resmi, bobot otomatis di-download

model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    project="helmet_detection",
    name="yolov8_helmet",
    device="0"
)
```

```python
import os
from ultralytics import YOLO
from IPython.display import Image, display

# Muat model hasil training
model = YOLO("/content/helmet_detection/yolov8_helmet/weights/best.pt")

# Folder validasi
val_dir = "/content/dataset/valid/images"

# Jalankan deteksi pada 5 gambar pertama
for img_name in os.listdir(val_dir)[:5]:
    img_path = os.path.join(val_dir, img_name)
    results = model(img_path)

    # Simpan hasil prediksi ke file baru
    save_path = f"/content/predict_{img_name}"
    results[0].save(filename=save_path)

    # Tampilkan hasil
    display(Image(filename=save_path, width=500))

```

```python
model.val()
```
