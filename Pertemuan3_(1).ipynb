```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
#@title Ambil data

import torchvision

data= torchvision.datasets.FashionMNIST(
    root='./data', train=True, download=True,
    )

# preposessing data ke numpy array
gambars = data.data.numpy().astype(float)
targets = data.targets.numpy()
n = len(gambars)
print(n)
```

```python
dict_class = {
    i:class_name for i, class_name in enumerate(data.classes)
}
labels = np.array([dict_class[y] for y in targets])

print("kelas : {}".format(dict_class))
# dimensi gambar
print("Dimensi gambar :{}".format(gambars[0].shape))
print("Jumlah gambar :{}".format(len(gambars)))
print("Type gambar :{}".format(type(gambars[0])))
print("Gambar : \n",gambars[0])
```
